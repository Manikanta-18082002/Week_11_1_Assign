Q1. What is the Probability density function?

### Probability Density Function (PDF)
is a function that describes the likelihood of a continuous random variable taking on a particular value within a given range. The PDF provides a way to model the distribution of the random variable's values over its entire possible range.

The Probability Density Function is typically denoted as f(x), where "x" represents the value of the random variable. The PDF satisfies the following properties:

The PDF is non-negative for all values of x: f(x) ≥ 0.
The total area under the PDF curve over the entire range of possible values is equal to 1.

Q2. What are the types of Probability distribution?

There are several types of probability distributions in statistics, each of which is used to model different types of random variables

#### Normal Distribution (Gaussian Distribution):
* Symmetrical and bell-shaped.
* Characterized by mean (μ) and standard deviation (σ).
* Many natural phenomena follow this distribution.
* Central limit theorem relies on it.

#### Binomial Distribution:
* Discrete distribution.
* Models the number of successes (binary outcome) in a fixed number of independent trials.
* Parameters: number of trials (n) and probability of success (p).

#### Poisson Distribution:
* Discrete distribution.
* Models the number of events occurring in a fixed interval of time or space.
* Parameter: average rate (λ) of events.

#### Uniform Distribution:
* Continuous distribution.
* All values within an interval are equally likely.
* Parameters: lower bound (a) and upper bound (b).

#### Exponential Distribution:
* Continuous distribution.
* Models the time between events in a Poisson process.
* Parameter: rate (λ) of event occurrences.

#### Log-Normal Distribution:
* Continuous distribution.
* Used to model variables that are the product of many small, independent factors.
* Logarithm of the variable follows a normal distribution.

#### Bernoulli Distribution:
* Discrete distribution.
* Models a binary outcome (e.g., success/failure) with a single trial.
* Parameter: probability of success (p).

Q3. Write a Python function to calculate the probability density function of a normal distribution with 
given mean and standard deviation at a given point

```python
import math

def normal_pdf(mean, std_dev, x):
    exponent = -((x - mean) ** 2) / (2 * std_dev ** 2)
    pdf = (1 / (std_dev * math.sqrt(2 * math.pi))) * math.exp(exponent)
    return pdf

# Example usage:
mean = 3.0
std_dev = 1.5
x = 4.0
pdf_value = normal_pdf(mean, std_dev, x)
print(f"PDF at x = {x}: {pdf_value:.4f}")
```

Q4. What are the properties of Binomial distribution? Give two examples of events where binomial distribution can be applied


The properties of a Binomial distribution are:

### Discrete Distribution:
The binomial distribution is a discrete probability distribution, meaning it models events with a finite number of distinct outcomes (usually two: success and failure).

### Binary Outcomes: 
It is used to model situations where each trial results in one of two possible outcomes, typically denoted as "success" and "failure."

### Fixed Number of Trials:
The distribution is defined for a fixed number of independent and identical trials, denoted as "n."

### Constant Probability of Success:
The probability of success (usually denoted as "p") remains constant from trial to trial. This probability is the same for each trial.

### Independent Trials:
Each trial is independent of the others, meaning the outcome of one trial does not affect the outcome of subsequent trials.

### Two Parameters:
The distribution is characterized by two parameters: "n" (the number of trials) and "p" (the probability of success on each trial).

Examples:

##### Coin Flips: 
Modeling the number of heads obtained when flipping a fair coin "n" times. Each flip is independent, with a fixed probability of success (getting a head) of 0.5.

##### Quality Control:
In a manufacturing process, you can use the binomial distribution to model the number of defective items in a batch of "n" items. Each item is inspected independently, and there is a fixed probability of each item being defective.

The binomial distribution is widely used in various fields, including statistics, quality control, biology, and social sciences, to model events that involve binary outcomes and repeated trials.

Q5. Generate a random sample of size 1000 from a binomial distribution with probability of success 0.4  and plot a histogram of the results using matplotlib.

```python
import numpy as np
import matplotlib.pyplot as plt

# Setting the parameters
n = 1000  # Sample size
p = 0.4   # Probability of success

# Generating the random sample from the binomial distribution
random_sample = np.random.binomial(n, p, size=n)

# Create a histogram
plt.hist(random_sample, bins=20, density=True, color='skyblue', alpha=0.7)
plt.title('Histogram of Binomial Distribution')
plt.xlabel('Number of Successes')
plt.ylabel('Probability Density')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

Q6. Write a Python function to calculate the cumulative distribution function of a Poisson distribution with given mean at a given point.

```python
import math

def poisson_cdf(mean, x):
   
    if x < 0:
        return 0.0  # CDF is 0 for negative values
    cdf = 0.0
    for k in range(x + 1):
        cdf += math.exp(-mean) * (mean ** k) / math.factorial(k)
    return cdf

mean = 3.0 #Ex:
x = 2
cdf_value = poisson_cdf(mean, x)
print(f"CDF at x = {x} is {cdf_value:.4f}")
```

Q8. Generate a random sample of size 1000 from a Poisson distribution with mean 5 and calculate the sample mean and variance.

```python
import numpy as np

# Setting the parameters
mean = 5  # Mean of the Poisson distribution
sample_size = 1000

# Generating the random sample from the Poisson distribution
random_sample = np.random.poisson(mean, size=sample_size)

# Calculating the sample mean and variance
sample_mean = np.mean(random_sample)
sample_variance = np.var(random_sample)

print(f"Sample Mean: {sample_mean:.4f}")
print(f"Sample Variance: {sample_variance:.4f}")
```

Q9. How mean and variance are related in Binomial distribution and Poisson distribution?

In both the Binomial distribution and the Poisson distribution, there is a relationship between the mean (μ) and the variance (σ²) of the distribution, but the nature of this relationship differs between the two distributions.

### Binomial Distribution:
* Here we have a fixed number of independent trials (n) with a constant probability of success (p) in each trial, the relationship between the mean and the variance is as follows:
* Mean (μ) = n * p
* Variance (σ²) = n * p * (1 - p)
The variance in a Binomial distribution is influenced by both the number of trials (n) and the probability of success (p). As the number of trials or the probability of success changes, it affects the spread or variability of the distribution.

### Poisson Distribution:
* In a Poisson distribution, which models the number of events occurring in a fixed interval of time or space, the relationship between the mean (λ, also called the rate parameter) and the variance (σ²) is simpler:
* Mean (λ) = Variance (σ²)
In the Poisson distribution, the mean and variance are equal. This means that the variability in a Poisson distribution is solely determined by the mean rate parameter. If the mean rate increases, the spread or variability of the distribution also increases.

In short: the key difference in the relationship between mean and variance in these two distributions is that in the Binomial distribution, the variance depends on both the number of trials and the probability of success, while in the Poisson distribution, the variance is solely determined by the mean rate parameter, and they are equal.






Q10. In normal distribution with respect to mean position, where does the least frequent data appear?

In a normal distribution, the least frequent data appears in the tails of the distribution, farthest away from the mean.

A normal distribution is a symmetric, bell-shaped curve, and the majority of the data points are concentrated around the mean. As you move away from the mean in either direction (left or right), the frequency of data points decreases. The tails of the distribution represent the extreme values or outliers, and they occur less frequently.

* The data points to the left of the mean (in the left tail) are the least frequent and represent values that are significantly below the mean.

* The data points to the right of the mean (in the right tail) are also the least frequent and represent values that are significantly above the mean.
